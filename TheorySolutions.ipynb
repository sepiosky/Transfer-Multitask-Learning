## سوال ۱: بهینە سازی در یادگیری چند وظیفه ای

الف)

در گذشته دلایل ضعف بهینه سازی یادگیری چند وظیفه ای، نرخ متفاوت یادگیری وظیفه ها و یا ضعف های ساختاری در نظر گرفته می شده است. اما یک دلیل که می تواند منطقی باشد تداخل بردار گرادیان وظایف مختلف (ضرب داخلی آنها منفی باشد) در زمان بهینه سازی است. اما تداخل گرادیان ها به تنهایی آنقدر مشکل ساز نیست. با توجه به آبجکتیو یادگیری چندوظیفه ای که معمولا میانگین (شاید وزن دار) تابع هزینه است، میانگین گیری بردار های گرادیان به طور کلی میتواند حتی در حالتی که گرادیان ها تداخل دارند نیز آبجکتیو کلی یادگیری چندوظیفه ای را افزایش دهد. (گرچه منجر به کاهش در بعضی وظایف به صورت مقطعی نیز بشود) اما شرایطی وجود دارد که در کنار تداخل گرادیان ها می تواند به کاهش عملکرد کلی یادگیری روی وظایف مختلف بشود. اگر به طور مثال گرادیان های ۲ وظیفه با هم تداخل داشته باشند و یکی از انها از لحاظ اندازه به دیگری برتری قابل توجهی داشته باشد در میانگین گرفتن جهت کلی گرادیان به گرادیان وظیفه غالب متمایل می شود به این امید که بهینه سازی در جهت گرادیان وظیفه غالب با توجه با اندازه گرادیان باعث کاهش چشمگیر هزینه کلی می شود و افزایش هزینه وظیفه مغلوب قابل چشم پوشی است.اما اگر نقطه ای از منیفولد پارامتر ها که بهینه سازی در این لحظه در آن انجام می شود انحنا شدیدی داشته باشد ممکن است مشکلاتی بوجود بیاید. انحنا به صورت زیر تعریف می شود: 

$$ H(\mathcal{L}; \theta, \theta^{\prime}) = \int_{0}^{1} \mathcal{\nabla}\mathcal{L}(\theta)^T [\mathcal{\nabla}^2\mathcal{L} (\theta + \alpha(\theta^\prime - \theta))] \mathcal{\nabla}\mathcal{L}(\theta) d \alpha  $$

که انحنای میانگین تابع هزینه بین $\theta$ و $\theta^\prime$ در مسیر گرادیان آبجکتیو یادگیری چند وظیفه ای (میانگین گرادیان ها) است. اگر مقدار $H$ از یک حدی بیشتر باشد نشان دهنده انحنایی بالایی منیفولد پارامتر ها در مسیر بهینه سازی گرادیان فعلی است. حال اگر در موقعیتی که گرادیان ها تداخل دارند و غلبه چشم گیری در میان گرادیان ها وجود دارد، مسیر کلی گرادیان در یک نقطه از منیفولد با انحنای بالا گیر کرده باشد ممکن است سود حاصل از بهینه سازی گرادیان غالب بیش از حد واقعی تخمین زده شده باشد و گرادیان غالب به دلیل انحنای بالا نتواند هزینه را به خوبی کاهش دهد. از سوی دیگر گرادیان های مغلوب که احتمالا در جهت خلاف انحنای شدید منیفولد قرار دارند، ممکن است باعث افزایش در تابع هزینه به مقداری بیش از مقدار قابل پیش بینی منجر شوند. در این صورت آن تصمیم که سود حاصل از بهینه سازی در جهت گرادیان غالب را بیشتر از ضرر حرکت در خلاف جهت گرادیان مغلوب تخمین میزد به دلیل انحنای بالای آبجکتیو باعث متضرر شدنن آبجکتیو کلی یادگیری چند وظیفه ای می شود.

به عنوان مثال، در شکل بالا مشاهده می شود نقطه بهینه بهینه سازی روی خط تلاقی دو دره است اما در شکل d بعد از چند قدم گرادیان، گرادیان تسک ۱ در مسیر بهینه قرار گرفته و از لحاظ اندازه کوچک می شود و مغلوب گرادیان تسک ۲ می شود (۱) همچنین به دلیل ماهیت پارامتر ها دو گرادیان متداخل اند (۲) و همچنین این تداخل با تفاوت اندازه در نقطه ای با انحنای بالا رخ داده است. در این حالت در جهت گرادیان تسک ۱ که مسیر بهینه است اما گرادیانش کوچک است حرکتی صورت نمیگیرد و همچنین به دلیل انحنای بالا در مسیر گرادیان تسک ۲ نیز حرکت صورت نمیگیرد و در حالت غیر بهینه کلی متوقف می شود.

پس عواملی که در صورت بروز همزمان مشکل زا می شوند عبارت اند از:
- تداخل گرادیان ها
- غالبیت بعضی از گرادیان ها از لحاظ اندازه
- انحن های بالا در تابع بهینه سازی به خصوص در نقاط تداخل

ب)

برای حل این مشکل، در مقاله صورت سوال راه حلی برای نقض یکی از این ۳ شرط خطرناک و جلوگیری از بروز همزمان آنها ارائه شده است. این راه حل روشی برای تغییر گرادیان ها بدون داشتن فرض خاصی راجع به شرایط بهینه سازی و یا مدل مساله ارائه میکند. این روش در زمان تداخل گرادیان ها آنها را به نحوی تغییر می دهد که اثرات منفی تداخل و پیش بینی دست بالا یا پایین در مسائلی با انحنای بالا در فضای بهینه سازی را تعدیل کند.

راه حل پیشنهادی به این صورت است که برای هر ۲ گرادیان متداخل (با ضرب داخلی منفی) هر کدام به صفحه نرمال بردار دیگر تصویر شود. به این شکل مولفه های متداخل حذف شده و گرادیان ها تقریبا هم جهت می شوند. این کار به صورت متناوب روی بچ ها انجام می شود تا تداخل گرادیان ها به حداقل برسد

بررسی های تئوری راه حل در مقاله ارائه شده است. ابتدا در حالت ۲ وظیفه ای و بهینه سازی محدب همگرایی روش به راه حل بهینه بررسی شده است که به جز مواردی که گرادیان ها کاملا متداخل اند (ضرب داخلی آنها -۱ و گرادیان های تصویر شده صفر میشوند) به جواب بهینه همگرا می شود. حالت گرادیان های کاملا متناقض با توجه به داشتن معمولا بیش از ۲ وظیفه خاصیت تصادفی الگوریتم های بهینه سازی متداول زیاد محتمل نیستند. در بررسی های تئوری محاسباتی برای شرایط ۳ گانه خطرناک انجام شده است تا نشان دهد برای چه حد از تداخل گرادیان و غلبه اندازه ای و انحنای فضای پارامتر های روش ارائه شده نسبت به روش های بهینه سازی عادی دارای ارجحیت هستند. همچنین علاوه بر ۳ شرط گفته شده یک شرط دیگر بررسی شده است و آن هم __بزرگ بودن کافی نرخ یادگیری__ است زیرا بزرگ بودن نرخ یادگیری در مجاورت انحنای شدید منیفولد باعث خوشبینی در بهینه بودن میاگین گرادیان های متداخل می شود. کارایی راه حل در مسائل بزرگ یادگیری چند وظیفه ای روی دیتاست های یادگیری با نظارت و محیط های یادگیری تقویتی در مقاله بررسی شده اند.

## سوال ۲: یادگیری چند وظیفە ای

الف)

با توجه به مساله یک فرم مناسب از مساله برنامه ریزی مرتبه ۲ برای این مساله به شکل زیر است:

$$
\begin{gather}
    \min \frac{1}{2} x^{T}Cx + p^Tx                       \nonumber\\
    \begin{aligned}
        \text{s.t.} \quad   Ax \geq b           \nonumber\\
                                   \label{eqn:nonnegative}
    \end{aligned}
\end{gather}
$$

مساله بهینه سازی مورد نظر در این قالب به شکل زیر در می آید:

$$
\begin{gather}
    \min_{g^{\prime}} {\lVert g^{\prime} - g  \rVert }_{2}^{2} =  \min_{g^{\prime}} {g^{\prime}}^Tg^{\prime} + g^Tg -2{g^{\prime}}^Tg =  \min_{g^{\prime}} {g^{\prime}}^Tg^{\prime} -2g^Tg^{\prime} = \min_{x} \frac{1}{2} x^{T}Cx + p^Tx \nonumber\\
    \begin{aligned}
        \text{where} \ \ \ \ \ \ \              C = 2I_{n}, \ \ \  p = -2g, \ \ \ x=g^{\prime}           \nonumber\\
                                   \label{eqn:nonnegative}
    \end{aligned}
\end{gather}
$$

$$
\begin{gather}
    \begin{aligned}
            \text{s.t.}  \ \ \ \langle g^{\prime},g_{k}\rangle \geq 0  \ \ \ \ \            \forall k<t           \nonumber\\
                                   \label{eqn:nonnegative}
    \end{aligned}
\end{gather}
$$

که معادل قید زیر است

$$
\begin{gather}
    \begin{aligned}
            \text{s.t.}  \ \ \   Ax>b \ \ \     \text{where} \ \ \  A = \begin{bmatrix}
            \text{---} & g_1 & \text{---} \\
            \text{---} & g_2 & \text{---} \\
                       & \vdots &  \\
            \text{---} & g_{t-1} & \text{---} 
        \end{bmatrix}, \ \ \ x=g^{\prime}, \ \ \ b=0          \nonumber\\
                                   \label{eqn:nonnegative}
    \end{aligned}
\end{gather}
$$

ب)

با توجه به انواع برنامه ریزی های اشاره شده در صفحه ۶ مقاله ذکر شده در صورت سوال، این مساله از نوع $II$ است و فرم دوگان برنامه ریز خطی قسمت الف به شکل زیر است:

$$
\begin{gather}
    \max_{u,v} -\frac{1}{2} u^{T}Cu + b^Tv                       \nonumber\\
    \begin{aligned}
        \text{s.t.} \quad   A^{T}v - Cu = p, \ \ \ v \geq 0           \nonumber\\
                                   \label{eqn:nonnegative}
    \end{aligned}
\end{gather}
$$

با توجه به اثبات فرم دوگان در سکشن ۵ و ۶ مقاله جواب بهینه $(u_0, v_0)$ مساله دوگان معادل جواب بهینه $x_0$ مساله اصلی است که $Cx_0 = Cu_0 $ و با توجه به همانی بودن C دو جواب $x_0$. و $u_0$ برابر هستند و با داشتن $u_0$ مقدار $v_0$ از حل یک مساله برنامه ریزی خطی بدست می آید. اما در اینجا برای رسیدن از جواب دوگان به جواب اصلی با جاگذاری $x$ به جای $u$ در معادله بالا و با توجه به نتایج قسمت الف برای متغیر های مختلف داریم:

$$
\begin{gather}
    \max_{g^{\prime}, v} -\frac{1}{2} {g^{\prime}}^{T}Cg^{\prime} + b^Tv = \max_{g^{\prime}} -\frac{1}{2} {g^{\prime}}^{T}(2I)g^{\prime} + 0*v =  \max_{g^{\prime}} - {g^{\prime}}^{T}g^{\prime}                      \nonumber\\
    \begin{aligned}
        \text{s.t.} \quad   A^{T}v - 2g^{\prime} = -2g \ \  \rightarrow  \ \ \ g^{\prime} = \frac{1}{2} A^{T}v + g,   \ \ \ v \geq 0         \nonumber\\
                                   \label{eqn:nonnegative}
    \end{aligned}
\end{gather}
$$

با جاگذاری قید بدست آمده در تابع داریم: 

$$
\begin{gather}
\max_{v} - {(\frac{1}{2}A^Tv+g)}^{T}{(\frac{1}{2}A^Tv+g)}    =  \max_{v}   -\frac{1}{4} v^TAA^Tv -  g^TA^Tv  = \min_{v}   v^TAA^Tv + 4g^TA^Tv  = \min_{v}   v^TKv + 4Gv            \nonumber\\
    \begin{aligned}
        \text{where} \ \ \ K =  \begin{bmatrix}
            \langle g_{1},g_{1}\rangle & \langle g_{1},g_{2}\rangle  & \text{---} & \langle g_{1},g_{t-1}\rangle \\
            \langle g_{2},g_{1}\rangle & \langle g_{2},g_{2}\rangle & \text{---} & \langle g_{2},g_{t-1}\rangle \\
                       & \vdots &  \\
            \langle g_{t-1},g_{1}\rangle & \langle g_{t-1},g_{2}\rangle & \text{---} & \langle g_{t-1},g_{t-1}\rangle 
        \end{bmatrix} \ \ \ and \ \ \ G =  \begin{bmatrix}
            \langle g,g_{1}\rangle & \langle g,g_{2}\rangle  & \text{---} & \langle g,g_{t-1}\rangle \\
        \end{bmatrix} , \ \ \ v \geq 0      \nonumber\\
                                   \label{eqn:nonnegative}
    \end{aligned}
\end{gather}
$$

با حل مساله دوگان برای $v$ و جاگذاری جواب در معادله $g^{\prime} = \frac{1}{2} A^{T}v_0 + g$ می توان گرادیان مقید شده $g^{\prime}$ که بیشترین نزدیکی به گرادیان تسک جدید ($g$) را با حفظ بهینگی تابع رو داده ی کش تسک های گذشته دارد
($\  \langle g^{\prime},g_{k}\rangle \geq 0  \ \ \forall k<t$)     پیدا کرد. 

ج)

متغیر بهینه سازی فرم اولیه $g^{\prime}$ است که گرادیان پارامتر های مدل رو تسک جدید است. در نتیجه فرم اولیه یک برنامه ریزی مرتبه ۲ با $n$ متغیر است که $n$ که پارامتر های شبکه است و می تواند از چند صد هزار تا چند صد میلیون باشد اما فرم دوگان مساله یک بزنامه ریزی مرتبه ۲ با متغیر $v$  که یک بردار $t-1$ تایی است که $t$ تعداد تسک هایی است که مدل برای آن آموزش داده می شود که معمولا چند ده تا است در نتیجه حل مساله دوگان به دلیل تعداد متغیر های بسیار کمتر عملیاتی تر است.

$$

$$

    x \\
    y \\
\end{pmatrix})

$$
